## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

### My Week 1 discovery (Option B)

**Research question:** Does CTR differ by `content_type` when pages are in the same `position_tier`?

**Why it matters:** If one content type gets fewer clicks at similar rankings, refresh priorities might differ by format — as decision-support only, not proof of Google's algorithm.


```python
# Starting from the same df loaded earlier in the notebook
visible = df[
    (df["impressions_90d"] >= 100) &
    (df["position_tier"] != "no_data")
].copy()

print(f"Rows used for comparison: {len(visible):,}")
print("\nContent types:")
print(visible["content_type"].value_counts())

print("\nPosition tiers:")
print(visible["position_tier"].value_counts())
```

```python
ctr_by_group = (
    visible
    .groupby(["position_tier", "content_type"], as_index=False)
    .agg(
        mean_ctr=("ctr", "mean"),
        page_count=("ctr", "size")
    )
    .sort_values(["position_tier", "mean_ctr"])
)

ctr_by_group["mean_ctr"] = ctr_by_group["mean_ctr"].round(4)
ctr_by_group
```

```python
worst_per_tier = (
    ctr_by_group
    .sort_values(["position_tier", "mean_ctr", "page_count"], ascending=[True, True, False])
    .groupby("position_tier", as_index=False)
    .first()
)

worst_per_tier = worst_per_tier.rename(columns={
    "content_type": "lowest_ctr_content_type",
    "mean_ctr": "lowest_mean_ctr",
    "page_count": "pages_in_group"
})

worst_per_tier
```

```python
for _, row in worst_per_tier.iterrows():
    print(
        f"{row['position_tier']:10s} -> "
        f"{row['lowest_ctr_content_type']:18s} "
        f"CTR={row['lowest_mean_ctr']:.4f} "
        f"(n={row['pages_in_group']})"
    )
```

```python
small_groups = ctr_by_group[ctr_by_group["page_count"] < 30]
print(f"Groups with fewer than 30 pages: {len(small_groups)}")

if len(small_groups) > 0:
    display(small_groups.sort_values("page_count"))
else:
    print("All groups have at least 30 pages.")
```

```python
import matplotlib.pyplot as plt

pivot = visible.pivot_table(
    index="position_tier",
    columns="content_type",
    values="ctr",
    aggfunc="mean"
)

pivot.plot(
    kind="bar",
    figsize=(10, 5),
    title="Mean CTR by content type within each position tier"
)
plt.ylabel("Mean CTR (%)")
plt.xlabel("Position tier")
plt.xticks(rotation=45)
plt.legend(title="Content type", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

### What I observed

In this anonymized sample, I filtered to pages with `impressions_90d >= 100` and valid `position_tier`, then compared mean CTR by `content_type` within each tier.

- In **[tier 1]**, **[content_type A]** had the lowest mean CTR (**[value]**%, n=**[count]**).
- In **[tier 2]**, **[content_type B]** had the lowest mean CTR (**[value]**%, n=**[count]**).
- Overall pattern: **[one content type was lowest in X of Y tiers / differences varied by tier / differences were small]**.

**Directional takeaway:** Content format may influence click-through at similar ranking positions in this sample, but this is observed evidence only — not proof of Google's ranking behavior.

**Limitations:** Single 30k-row teaching sample; some groups may be small; comparison only, not causal.
